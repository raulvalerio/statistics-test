### Creating subsets by containing words

```python
fastfood_burger = fastfood[ fastfood['item'].str.contains('Burger|burger')]
fastfood_burger.head()
```

```python
fastfood_burrito = fastfood[ fastfood['item'].str.contains('Burrito')]
fastfood_burrito.head()
```

```python
fastfood_sandwich = fastfood[ fastfood['item'].str.contains('Sandwich')]
fastfood_sandwich.head()
```

### Feature Engineering

```python
fastfood['burger'] =  fastfood['item'].apply(lambda x: 1 if 'urger' in x else 0)

fastfood['my_groups'] =  fastfood['item'].apply(lambda x: 1 if 'urger' in x else 2 if 'Sandwich' in x else 0)
fastfood.head()
```

# CONFIRMATORY DATA ANALYSIS

Hypothesis based in observations from previous graphs

\begin{align}
H_0: \mu_1  = \mu_2 \\
H_1: \mu_1 \neq \mu_2 \\
\end{align}

### Calories in sandwiches: two independent samples comparison

```python
from scipy import stats

## x1 comes from previous lines: x1 is fastfood_sandwich

df1= x1.calories[x1.restaurant=='Burger King']
df2= x1.calories[x1.restaurant=='Mcdonalds']
                 
tprueba,pvalue =stats.ttest_ind(df1,df2, equal_var=False)
print("t-value: ",tprueba)
print("pvalue: ", pvalue)
```

How do we read this output?

### calories in burgers: two independent samples comparison

```python
x1= fastfood_burger
df1= x1.calories[x1.restaurant=='Burger King']
df2= x1.calories[x1.restaurant=='Mcdonalds']
                 
tprueba,pvalue =stats.ttest_ind(df1,df2, equal_var=False)
print("t-value: ",tprueba)
print("pvalue: ", pvalue)
```

### ANOVA
